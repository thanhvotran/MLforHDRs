# Kmean in Python, a walkthrough

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
# Generate 'fake' data

from sklearn.datasets import make_blobs
X, y_true = make_blobs(n_samples=300, centers=4,
                       cluster_std=0.60, random_state=0)
plt.scatter(X[:, 0], X[:, 1], s=50);
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200);
```

# Customer Segmentation in Marketing
This example uses the 'Mall Customers' dataset, which is popular for demonstrating customer segmentation.

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Load dataset
url = "https://raw.githubusercontent.com/tirthajyoti/Machine-Learning-with-Python/master/Datasets/Mall_Customers.csv"
data = pd.read_csv(url)

# Selecting relevant features for clustering
X = data[['Annual Income (k$)', 'Spending Score (1-100)']]

# Visualize the data before clustering
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(X['Annual Income (k$)'], X['Spending Score (1-100)'])
plt.title('Before Clustering')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')

# Standardizing the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Applying KMeans
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(X_scaled)
labels = kmeans.labels_

# Plotting the clusters after KMeans
plt.subplot(1, 2, 2)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels, cmap='viridis')
plt.title('After Clustering')
plt.xlabel('Annual Income (Standardized)')
plt.ylabel('Spending Score (Standardized)')
plt.colorbar(label='Cluster Label')

plt.tight_layout()
plt.show()
```

# Gene Expression Data in Biology
This example uses a synthetic dataset simulating gene expression data to illustrate clustering of different cell types based on gene expression patterns.

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Generate synthetic data resembling gene expression profiles
X, _ = make_blobs(n_samples=300, centers=4, random_state=42, cluster_std=1.5)

# Visualize the data before clustering
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1])
plt.title('Before Clustering')
plt.xlabel('Expression Level of Gene 1')
plt.ylabel('Expression Level of Gene 2')

# Applying KMeans to find 4 distinct cell types or conditions
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_

# Plotting the clusters to visualize different gene expression patterns after clustering
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.title('After Clustering')
plt.xlabel('Expression Level of Gene 1')
plt.ylabel('Expression Level of Gene 2')
plt.colorbar(label='Cluster Label')

plt.tight_layout()
plt.show()
```
